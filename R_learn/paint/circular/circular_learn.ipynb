### customize chromosome track

```r
set.seed(123)
circos.initializeWithIdeogram(plotType=NULL)
circos.track(ylim=c(0,1), panel.fun=function(x,y){
    chr = CELL_META$sector.index
    # chr 当前染色体名称
    xlim = CELL_META$xlim
    ylim = CELL_META$ylim
    # xlim:当前染色体的大小范围，xlim[1]=0;xlim[2]=max
    # ylim=(0, 1) track 高度
    circos.rect(xlim[1], 0, xlim[2], 1, col=rand_color(1))
    # rect 矩阵绘图
    circos.text(mean(xlim), mean(ylim), chr, cex=0.7, col="white",
                facing="inside", ,niceFacing=TRUE)
    # facing:文字朝向，niceFacing:是否根据视觉效果进行调整
}, track.height=0.2, bg.border=NA)# bg.border: 是否需要外围框线
# track.height： track占圆半径的比例
```

## modes for circos.genomicTrack()

### normal mode

```r
# data frame
data =generateRandomBed(nc=2)
circos.genomicTrack(data, numeric.column=4,
        panel.fun = function(region,value,...){
            circos.genomicPoints(region, value, ...)
            circos.genomicPoints(region,value)
            # 1st column in `value` while 4th column in `data`
            circos.genomicPoints(region, value, numeric.column=1)
        })
```

```r
# list of data frame
#If it is not specified, the first numeric column in each data frame is used
bed_list = list(generateRandomBed(), generateRandomBed())
circos.genomicTrack(bed_list,
    panel.fun=function(region,value,...){
        i=getI(...)#The index for the current data frame can be get by getI(...). Note getI(...) can only be used inside panel.fun and ... argument is mandatory.
        circos.genomicPoints(region,value,col=i,...)
    })
```

```r
# column 4 in the first bed and column 5 in the second bed
circos.genomicTrack(bed_list, numeric.colum=c(4, 5),
        panel.fun=function(region, value,...){
            i=getI(...)
            circos.genomicPoints(region, value, col=1, ...)
        })
```

### stack mode


```r
# data frame
# ylim is re-defined to c(0.5, n+0.5) in which n is number of numeric columns specified. panel  
# The value of the index of “current” numeric column can be obtained by getI(...)  
data = generateRandomBed(nc=2)
circos.genomicTrack(data, stack=TRUE,
        panel.fun=function(region, value,...){
            i=getI(...)
            circos.genomicPoints(region, value, col=i, ...)
        })
```

```r
#W hen input data is a list of data frames, each data frame will be treated as a single unit. 
# ylim is re-defined to c(0.5, n+0.5) in which n is the number of data frames
bed_list = list(generateRandomBed(), generateRandomBed())
circos.genomicTrack(bed_list, stack=TRUE,
        panel.fun=function(region, value, ...){
            i=getI(...)
            circos.genomicPoints(region, value, ...)
        })
```

```r
#In track B, the input is a list of data frames. Under stack mode, each data frame is added to a horizontal line
#Under stack mode, by default, the height of rectangles is internally set to make them completely fill the cell in the vertical direction. ytop and ybottom can be used to adjust the height of rectangles
circos.par("track.height" = 0.15, start.degree=90,
          canvas.xlim=c(0,1), canvas.ylim=c(0,1),gap.degree=270)
circos.initializeWithIdeogram(chromosome.index="chr1", plotType=NULL)
col_fun = colorRamp2(breaks = c(-1, 0, 1), colors=c("green", "black", "red"))
bed1 = generateRandomBed(nr=100)
bed2 = generateRandomBed(nr=100)
bed_list = list(bed1, bed2)
circos.genomicTrack(bed_list, stack=TRUE,
        panel.fun = function(region, value,...){
            i=getI(...)
            circos.genomicRect(region, value, ytop=i+0.3, ybottom=i-0.3,
                    col=col_fun(value[[1]]), ...)
        })
circos.clear()
```

```r
# In track D, rectangles are used to make barplots. We specify the position of the top of bars by ytop.column
circos.par("track.height" = 0.15, start.degree=90,
          canvas.xlim=c(0,1), canvas.ylim=c(0,1),gap.degree=270)
circos.initializeWithIdeogram(chromosome.index="chr1", plotType=NULL)
col_fun = colorRamp2(breaks = c(-1, 0, 1), colors=c("green", "black", "red"))
bed = generateRandomBed(nr=200)
circos.genomicTrack(bed,
        panel.fun = function(region, value, ...){
            circos.genomicRect(region, value, ytop.column=1, ybottom=0,
                    col=ifelse(value[[1]]>0, "red", "green"), ...)
            circos.lines(CELL_META$cell.xlim, c(0, 0), lty=2, col="#00000040")
        })
circos.clear()
```
