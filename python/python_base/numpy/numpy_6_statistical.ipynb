生成正态分布数据

```python
import numpy as np
```

```python
mu, sigma = 0.0, 1.0
```

```python
g = np.random.normal(loc=mu, scale=sigma, size=1000)
g
```

```python
g1 = np.random.randn(1000)
g1
```

```python
g = (g - g.min()) / 10
phy = np.round(((100.1 - 20) * g + 20), 1)
g1 = (g1 - g1.min()) / 10
mat = np.round(((100.1 - 20) * g + 20), 1)
```

```python
phy
```

```python
mat
```

```python
np.mean(phy)#平均数
```

```python
np.std(phy)#标准差
```

```python
np.min(phy), np.max(phy)
```

```python
np.percentile(phy, 80)#百分位数
```

```python
np.median(phy)#中位数
```

```python
phy[phy > np.mean(phy)]
```

```python
np.where(phy > mat, 1, -1)#大于则返回1， 否则返回-1
```

```python
np.where(phy > phy.mean())#返回满足条件的索引
```

```python
a = np.array([1, -2, 1])#降序排列多项式系数
p = np.poly1d(a)
p# x**2 - 2x + 1 
```

```python
p(1)# 1**2 - 2*1 + 1
```
