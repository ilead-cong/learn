# 曲线图和散点图

## 曲线图
plt.plot()  
1.色彩：color  
2.线：linestyle/linewidth  
3.数据点：marker/markersize/markerfacecolor  

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### 曲线颜色设置

```python
x = np.linspace(0, 2*np.pi, 100)
plt.plot(x, np.sin(x), color='blue')#单词
plt.plot(x, np.sin(x-np.pi/3), color='r')#简写
plt.plot(x, np.cos(x), color='#A52A2A')#数字
```

### 曲线线形

```python
plt.plot(x, 2*x, linestyle='dashdot')#单词
plt.plot(x, -2*x + 2, linestyle=":")#符号
```

名称    solid    dashed    dashdot    dotted
符号    -       --       -.       :
样式   _____    -------   ..-.-.-..   ......

```python
plt.plot(x, np.sin(x), linestyle='-.', color='blue')
plt.plot(x, np.cos(x), linestyle=':', color='black')
```

### 数据点设置

```python
plt.plot(range(10), linestyle='--', marker='o', 
        markersize = 16, markerfacecolor='b', color='r')#marker设置点的形状， markersize设置点的大小， markerfacecolor设置点的颜色
plt.grid(True)
```

```python
#同样的效果
plt.plot(range(10), '--or', markersize=16, markerfacecolor='b')#--or, --线形，o点的形状， r线的颜色
plt.grid(True)
```

```python
plt.plot(range(10), '-Dr', markersize=16, markerfacecolor='b', 
        markevery=[2, 4, 6])# D表示钻石型，makervery表示需要标记的数据点，第二个，第四个，和第六个（数据点的顺序从第0个开始）
plt.grid(True)
```

```python
x = np.linspace(-np.pi, np.pi, 9)
plt.plot(x, np.sin(x), 'Dr', markersize=16, markerfacecolor='b', 
        markevery=[2, 4, 6])#Dr,未标明线形，那么只画出相应的点
plt.plot(x, np.cos(x), 'hr', markersize=16, markerfacecolor='m')
plt.grid(True)
```

### 曲线宽度

```python
x = np.linspace(-np.pi, np.pi, 9)
plt.plot(x, np.cos(x), linewidth=8)#linewidth线的宽度
```

### 图例

plt.legend()
    参数：loc的值
名称    best    upper right    upper left    lower left 
编号    0      1           2          3
名称    lower right    right    center left    center right
编号    4           5      6           7
名称    lower center    upper center    center
编号    8            9           10

```python
a = np.arange(0, 3, 0.02)#间隔为0.02
b = np.arange(0, 3, 0.02)
c = np.exp(a)#返回e的a次方
d = c[: : -1]

plt.plot(a, c, 'k--', label='Model')#label标签名称
plt.plot(a, d, 'r:', label='Data')
plt.plot(a, c+d, 'b-', label='Total')
plt.legend(loc='upper right')#loc标签位置
```

#### 绘制某条线的图例

```python
line1, = plt.plot(a, c, 'k--', label='model')
line2 = plt.plot(a, d, 'r:', label='data')[0]
line3, = plt.plot(a, c+d, 'b-', label='total')
plt.legend(handles=(line1, line2), loc=0)#此时line是一个多为对象，包括color，label等，需要将句柄handles作为句柄传递给legend()
```

## 散点图
plt.scatter

```python
rng = np.random.RandomState(0)#伪随机数， 种子是0
x = rng.randn(100)#以给定的形状创建一个数组，素组元素来符合标准正态分布N(0, 1)
y = rng.randn(100)
colors = rng.rand(100)#randn以给定的形状创建一个数组，并在数组中加入在[0, 1]之间均匀分布的随机样本
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)#alpha表示透明度

plt.colorbar()#绘制颜色图例
```
