## swarmplot
反应数据的分布情况

```python
import matplotlib.pyplot as plt

planets = sns.load_dataset('planets')

f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale('log')

sns.boxplot(x='distance', y='method', data=planets, whis=np.inf, palette='vlag')#whis,控制在超过高低四分位数时IQR的比例，因此需要延长绘制的触须线段。超出此范围的点将被识别为异常值;palette,调色板名称


sns.swarmplot(x='distance', y='method', data=planets, size=2, color='0.3', linewidth=0)

ax.xaxis.grid(True)
ax.set(ylabel='')
sns.despine(trim=True, left=True)
```

## 提琴图

```python
rs = np.random.RandomState(0)
n, p = 40, 8
d = rs.normal(0, 2, (n, p))
d += np.log(np.arange(1, p+1)) * -5 + 10#np.log，计算自然对数

sns.violinplot(data=d, inner="points")#inner，提琴图内部的表示
```
