## swarmplot


```python
import matplotlib.pyplot as plt

planets = sns.load_dataset('planets')

f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale('log')

sns.boxplot(x='distance', y='method', data=planets, whis=np.inf, palette='vlag')#whis,IQR;palette,


sns.swarmplot(x='distance', y='method', data=planets, size=2, color='0.3', linewidth=0)

ax.xaxis.grid(True)
ax.set(ylabel='')
sns.despine(trim=True, left=True)
```
