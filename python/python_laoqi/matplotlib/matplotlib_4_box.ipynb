```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## 普通箱线图

```python
fig, ax = plt.subplots(1, 2)#subplot(numbRow, numbCol, plotNum), numbRow是plot图的行数， numbCol是plot图的列数，plotNum是指第几行第几列的第几幅图
data = [1, 5, 9, 2]

#子图1
ax[0].boxplot([data])
ax[0].grid(True)
```

```python
fig, ax = plt.subplots(1, 2)
data = [1, 5, 9, 2]

#子图1
ax[0].boxplot([data])
ax[0].grid(True)

#子图2
ax[1].boxplot([data], showmeans=True) #显示平均值
ax[1].grid(True)
```

## 凹槽箱线图

```python
#notch:默认为None, 如果为True，则意味着绘制带有凹槽的箱线图
#凹槽代表数据分布的置信区间
plt.boxplot([2, 4, 6, 8, 10], notch=True)
plt.grid(True)
```

## 数据集箱线图

```python
import numpy as np
```

```python
np.random.seed(10)
collection_1 = np.random.normal(100, 10, 200)#生成一个正态分布的数据，loc=100，正态分布的均值为100；scale=10，正态分布的标准差为10，size=200，数据个数为200
collection_2 = np.random.normal(80, 30, 200)
collection_3 = np.random.normal(90, 20, 200)
collection_4 = np.random.normal(70, 25, 200)
data = [collection_1, collection_2, collection_3, collection_4]

fig = plt.figure(1, figsize=(9, 6))
bp=plt.boxplot(data)
```

```python
np.random.seed(10)
collection_1 = np.random.normal(100, 10, 200)#生成一个正态分布的数据，loc=100，正态分布的均值为100；scale=10，正态分布的标准差为10，size=200，数据个数为200
collection_2 = np.random.normal(80, 30, 200)
collection_3 = np.random.normal(90, 20, 200)
collection_4 = np.random.normal(70, 25, 200)
data = [collection_1, collection_2, collection_3, collection_4]

fig = plt.figure(1, figsize=(9, 6))
ax = plt.subplot(111)#subplot
bp=ax.boxplot(data, patch_artist=True)#箱体填充
```

```python
bp
```

## 箱线图属性

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)
```

### 修改矩形框内的填充色和矩形框的边线

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)

for box in bp['boxes']:
    box.set(color='#7570b3', linewidth=2)#设置边框颜色和粗细
    box.set(facecolor='#1b9e77')#填充色

```

### 设置须线的粗细和颜色

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)

for whisker in bp['whiskers']:
    whisker.set(color='red', linewidth=2)
```

### 设置上下限的线的颜色和粗细

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)

for cap in bp['caps']:
    cap.set(color='black', linewidth=2)
```

### 设置中位数的线的颜色和粗细

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)

for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=4)
```

### 设置离群值符号

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)

for flier in bp['fliers']:
    flier.set(marker='*', color='#e7298a', alpha=0.5)
```

### 设置坐标轴

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)

ax.set_xticklabels(['sample1', 'sample2', 'sample3', 'sample4'])#X轴下标
ax.get_xaxis().tick_bottom()#设置坐标轴标签的位置
ax.get_yaxis().tick_left()
#ax.get_xaxis().tick_top()
```

```python
fig, ax = plt.subplots(1, 2)
data_1 = [1, 5, 9, 2]

#子图1
ax[0].boxplot([data_1])
ax[0].grid(True)
ax[0].set_xticklabels(['sample1'])
ax[0].get_xaxis().tick_bottom()#设置坐标轴标签的位置
ax[0].get_yaxis().tick_left()


#子图2
ax[1].boxplot([data_1]) #显示平均值
ax[1].grid(True)
ax[1].set_xticklabels(['sample2'])
ax[1].get_xaxis().tick_top()#设置坐标轴标签的位置
ax[1].get_yaxis().tick_right()
```

### 整体效果

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)
```

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(data, patch_artist=True)

for box in bp['boxes']:
    box.set(color='#7570b3', linewidth=2)#设置边框颜色和粗细
    box.set(facecolor='#1b9e77')#填充色
    
for whisker in bp['whiskers']:
    whisker.set(color='red', linewidth=2)
    
for cap in bp['caps']:
    cap.set(color='black', linewidth=2)

for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=4)

for flier in bp['fliers']:
    flier.set(marker='*', color='#e7298a', alpha=0.5)

ax.set_xticklabels(['sample1', 'sample2', 'sample3', 'sample4'])#X轴下标
ax.get_xaxis().tick_bottom()#设置坐标轴标签的位置
ax.get_yaxis().tick_left()
```
