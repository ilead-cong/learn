# subplot


matplotlib一个figure对象可以包含多个子图(Axes)，可以使用subplot()绘制，subplot(numRows, numCols, plotNum)  
图表的整个绘图区域被分成numRows行和numCols列,plotNum参数指定创建的Axes对象所在的区域  
如果numRows, numCols和plotNum这三个数都小于10的话，可以把它们缩写成一个整数

## 位置调用

### 规则划分

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.figure()

#图1
plt.subplot(221)
data = [1, 5, 3, 2]
plt.boxplot([data])

#图2
plt.subplot(222)
data = [1, 5, 9, 2]
plt.boxplot([data])

#图3
plt.subplot(223)
data = [1, 5, 10, 2]
plt.boxplot([data])

#图4
plt.subplot(224)
data = [11, 5, 9, 2]
plt.boxplot([data])


plt.show()
```

### 不规则划分

```python
import numpy as np

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

#图1
plt.figure(12)
plt.subplot(221)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'r--')

#图2
plt.subplot(222)
plt.plot(t2, np.cos(2 * np.pi * t2), 'r--')

#图3
plt.subplot(212)
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])

plt.show()
```

### 元组调用

```python
fig, ax = plt.subplots(1, 2)# fig, (ax1, ax2) = plt.subplots(1, 2)
data = [1, 5, 9, 2]

#图1
ax[0].boxplot([data])

#图2
ax[1].boxplot([data])
ax[1].grid(True)

#图3
#ax[2].boxplot([data], showmeans=True)

#图4
#ax[3].boxplot([data], showmeans=True)
#ax[3].grid(True)

plt.show()
```

### 多图矩阵调用

```python
fig, ax = plt.subplots(2, 2)# fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
data = [1, 5, 9, 2]

#图1
ax[0][0].boxplot([data])

#图2
ax[0][1].boxplot([data])
ax[0][1].grid(True)

#图3
ax[1][0].boxplot([data], showmeans=True)

#图4
ax[1][1].boxplot([data], showmeans=True)
ax[1][1].grid(True)

plt.show()
```
