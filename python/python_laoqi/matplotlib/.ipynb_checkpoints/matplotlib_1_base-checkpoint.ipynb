## matplotlib
基本绘图思想和绘图方法

### matlab风格

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
plt.plot(x, y1)
```

```python
y2 = np.cos(x)
plt.plot(x, y2)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
```

### 面向对象绘图

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])# 0.1和0.1表示X轴和Y轴距离figure对象的距离，0.8和0.8表示X轴和Y轴长度占figure对象长度的80%
x = np.linspace(0, 2*np.pi, 100)
ax.plot(x, np.sin(x))
```

## 设置坐标系

### 坐标网格

grid()

```python
x = np.arange(0, 5, 0.02)
y = np.exp(-x) * np.cos(2*np.pi*x)
plt.plot(x, y)
plt.grid(color='blue')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.grid(color='red')
ax.plot(x, y)
```

### 坐标轴  
set_xlabel()  
set_ylabel()  
set_xlim()  
set_ylim()  

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.grid(color='gray')
ax.plot(x, y)

ax.set_xlabel('x axis')
ax.set_xlim((-2, 10))
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.grid(color='gray')
ax.plot(x, y)

ax.set_xlabel("x axis")
ax.set_xlim((0, 5))

ax.set_xticks(np.linspace(0, 5, 11))#设置坐标轴刻度

plt.show()
```
