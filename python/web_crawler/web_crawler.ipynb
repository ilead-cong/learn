```python
import requests
```

```python
from lxml import html
```

```python
url = "https://qiwsir.github.io/"
```

```python
page = requests.get(url).content.decode('utf-8')
```

```python
print(page)
```

```python
sel = html.fromstring(page)
```

```python
title = sel.xpath('//article/h2/a/text()')
```

```python
print(title)
```
