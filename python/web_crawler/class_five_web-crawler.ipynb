# 网络爬虫

## 定义
按照一定的规则，自动抓取万维网信息的程序或脚本  
例如：搜索引擎、竞品分析等

## 流程
发送请求：向对方服务器发送待抓取网站的链接  
返回响应：返回请求的网页源代码  
数据储存：利用正则表达式或解析式对源代码进行清洗，并将目标数据存储到本地或者数据库  

## 使用的包及函数
包：requests\re\bs4  
函数：  
requests.get --基于URL，发送网络请求  
re.findall --基于正则表达式，搜寻目标数据  
bs4.BeautifulSoup --对HTML源代码做解析，便于目标数据的拆解  

# 案例

## 案例1 动态网页

```python
import requests
import re

url = r"http://www.redbull.com.cn/about/branch"
response = requests.get(url)
response
```

```python
response.text
```

```python
re.findall("<h2>红牛广西分公司</h2>", response.text)
```

```python
re.findall("<h2>.*?</h2>", response.text)
```

```python
company = re.findall("<h2>(.*?)</h2>", response.text)
company
```

```python
add = re.findall("<p class=\'mapIco\'>(.*?)</p>", response.text)
add
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(response.text)
soup
```

```python
soup.findAll(name = 'p', attrs={'class':'mailIco'})
```

```python
for i in soup.findAll(name = 'p', attrs={'class':'mailIco'}):
    print(i.text)
```

```python
mail = [i.text for i in soup.findAll(name = 'p', attrs={'class':'mailIco'})]
```

```python
tel = [i.text for i in soup.findAll(name = 'p', attrs={'class':'telIco'})]
```

```python
import pandas as pd
```

```python
pd.DataFrame({'company':company, 'add':company, 'mail':mail, 'tel':tel})
```

## 案例2 静态网页

```python
url = r"https://sh.lianjia.com/ershoufang/pudong/pg1/"
response = requests.get(url)
response
```

```python
#反爬虫，请求头伪装
Headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'
}
```

```python
url = r"https://sh.lianjia.com/ershoufang/pudong/pg1/"
response = requests.get(url, headers = Headers)
response
```

```python
soup = bs4.BeautifulSoup(response.text)
soup
```

```python
soup.findAll(name = 'a', attrs={'data-el':'region'})
```

```python
name = [i.text.strip() for i in soup.findAll(name = 'a', attrs={'data-el':'region'})]
name
```

```python
[i.text for i in soup.findAll(name = 'div', attrs={'class':'houseInfo'})]
```

```python
Type = [i.text.split('|')[0].strip() for i in soup.findAll(name = 'div', attrs={'class':'houseInfo'})]
Type
```

```python
size = [float(i.text.split('|')[1].strip()[:-2]) for i in soup.findAll(name = 'div', attrs={'class':'houseInfo'})]
size
```

```python
direction = [i.text.split('|')[2].strip() for i in soup.findAll(name = 'div', attrs={'class':'houseInfo'})]
direction
```

```python
decoration = [i.text.split('|')[3].strip() for i in soup.findAll(name = 'div', attrs={'class':'houseInfo'})]
decoration
```

```python
flool = [i.text.split('|')[4].strip() for i in soup.findAll(name = 'div', attrs={'class':'houseInfo'})]
flool
```

```python
year = [i.text.split('|')[5].strip() for i in soup.findAll(name = 'div', attrs={'class':'houseInfo'})]
year
```

```python
total = [float(i.text[:-1]) for i in soup.findAll(name = 'div', attrs={'class':'totalPrice'})]
total
```

```python
price = [int(i.text[2:-4]) for i in soup.findAll(name = 'div', attrs={'class':'unitPrice'})]
price
```

```python
pd.DataFrame({'name':name, 'Type':Type, 'size':size, 'direction':direction, 'decoration':decoration, 'flool':flool, 'year':year, 'total':total, 'price':price})
```

## 案例3 动态网站

```python
url = r"http://tianqi.2345.com/t/wea_history/js/201910/58362_201910.js"
response = requests.get(url)
response
```

```python
data = re.findall("ymd:'([\d-]+)'", response.text)
high = re.findall("bWendu:'([\d-]+)℃'", response.text)
high
weather = re.findall("tianqi:'(.*?)'", response.text)
aqi = re.findall("aqiInfo:'(.*?)'", response.text)
```

```python
pd.DataFrame({'data':data, 'high':high, 'weather':weather, 'aqi':aqi})
```
